```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv(r'C:\Users\Jacques07\Downloads\Banknote-authentication-dataset-.csv')
data.head()
```

```python
data.describe()
```

```python
means = data.mean()
stds = data.std()

print("Means:\n", means)
print("\nStandard Deviations:\n", stds)
```

```python
# Pairplot to see feature relationships
pd.plotting.scatter_matrix(data, figsize=(10, 10))
plt.suptitle("Scatter Matrix of Banknote Features", y=1.02)
plt.show()

# Correlation heatmap
plt.imshow(data.corr(), cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.xticks(range(len(data.columns)), data.columns, rotation=45)
plt.yticks(range(len(data.columns)), data.columns)
plt.show()
```

```python
data.hist(figsize=(10,8))
plt.suptitle("Feature Distributions")
plt.show()
```
