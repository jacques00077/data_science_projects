```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Banknote-authentication-dataset-.csv')
v1 = data['V1']
v2 = data['V2']
v2.max()
```

```python
banknote = np.column_stack((v1,v2))
km_res = KMeans(n_clusters=2).fit(banknote)
clusters=km_res.cluster_centers_
plt.title('Banknote Authentication Dataset')
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(v1,v2)
plt.scatter(clusters[:,0],clusters[:,1],s=500)
plt.show()
```
